```python
import MNISTDataLoader
import os.path
import random
```

```python
input_path = 'input'
training_images_filepath = os.path.join(input_path, 'train-images.idx3-ubyte')
training_labels_filepath = os.path.join(input_path, 'train-labels.idx1-ubyte')
test_images_filepath = os.path.join(input_path, 't10k-images.idx3-ubyte')
test_labels_filepath = os.path.join(input_path, 't10k-labels.idx1-ubyte')
```

Load MNIST Dataset

```python
mnist_dataloader = MNISTDataLoader.MnistDataloader(training_images_filepath, training_labels_filepath, test_images_filepath, test_labels_filepath)
(x_train, y_train), (x_test, y_test) = mnist_dataloader.load_data()
```

Show random training and test images

```python
images_2_show = []
titles_2_show = []
for i in range(0, 10):
    r = random.randint(1, 60000)
    images_2_show.append(x_train[r])
    titles_2_show.append('training image [' + str(r) + '] = ' + str(y_train[r]))    

for i in range(0, 5):
    r = random.randint(1, 10000)
    images_2_show.append(x_test[r])        
    titles_2_show.append('test image [' + str(r) + '] = ' + str(y_test[r]))   
```

```python
MNISTDataLoader.show_images(images_2_show, titles_2_show)
```
